```
#Clases
class auto():

  def __init__ (self):

    self.__largoChasis=250
    self.__anchoChasis=100
    self.__ruedas=4
    self.__puertas=5
    self.__enmarcha=False

  def arrancar(self,arrancamos):
    self.__enmarcha = arrancamos

    if(self.__enmarcha):
      condicion=self.__chequeo()
      

    if(self.__enmarcha and condicion):
      return "El auto esta en marcha"

    elif(self.__enmarcha and condicion==False):
      return "Condicion del auto no favorable para arrancar"

    else:
      return "El auto esta detenido"

  def estado(self):
    print("El auto tiene ", self.__ruedas, "ruedas. ", "El auto tiene: ", self.__largoChasis,"Largo. ")

  def __chequeo(self):
    print("Realizando chequeo interno")

    self.combustible="ok"
    self.aceite="mal"
    self.puertas="cerradas"

    if(self.combustible=="ok" and self.aceite=="ok" and self.puertas=="cerradas"):
      return True

    else:
      return False
    

Auto1 = auto()
print(Auto1.arrancar(True))
print(Auto1.estado())


Auto2=auto()
print(Auto2.arrancar(False))
print(Auto2.estado())
```

```
#HerenciaSuper

class Persona():
  def __init__(self, nombre, edad, direccion):

    self.nombre=nombre
    self.edad=edad
    self.direccion=direccion
  
  def descripcion(self):
    print("Nombre:", self.nombre, "\nEdad:", self.edad, "\nDireccion:", self.direccion)

class Empleado(Persona):

  def __init__(self,sueldo,antiguedad, nombreEmpleado, edadEmpleado, direccionEmpleado):

    super().__init__(nombreEmpleado,edadEmpleado,direccionEmpleado)
    self.sueldo=sueldo
    self.antiguedad=antiguedad
  
  def descripcion(self):
    super().descripcion()
    print("Sueldo:", self.sueldo, "\nAntiguedad:", self.antiguedad)

Antonio=Empleado(1500,5,"Antonio", 25, "Chile")
Antonio.descripcion()
```
